```python
# UPLOAD YOUR CSV FILE HERE!
# This cell allows you to upload the CSV file generated by the main script.
# You MUST run this cell before proceeding to any data analysis below!

import pandas as pd
from google.colab import files
uploaded = files.upload()

df = pd.read_csv('products.csv')
```

```python
# Filter by website and price range
# This cell lets you choose a specific website and a price range to filter the data

import pandas as pd
from IPython.display import display
import ipywidgets as widgets


# Dropdown for website selection
site_dropdown = widgets.Dropdown(
    options=sorted(df['Website'].unique()),
    description='Website:',
    style={'description_width': 'initial'}
)

# Float input for price range
min_price_input = widgets.FloatText(
    value=0.0,
    description='Min price:',
    style={'description_width': 'initial'}
)

max_price_input = widgets.FloatText(
    value=99999.0,
    description='Max price:',
    style={'description_width': 'initial'}
)

# Button to apply filter
filter_button = widgets.Button(
    description='Apply Filter',
    button_style='success'
)

output = widgets.Output()

# Function to filter and display
def on_filter_button_clicked(b):
    with output:
        output.clear_output()
        filtered_df = df[
            (df['Website'] == site_dropdown.value) &
            (df['Price'] >= min_price_input.value) &
            (df['Price'] <= max_price_input.value)
        ].sort_values(by='Price')
        display(filtered_df.reset_index(drop=True))

filter_button.on_click(on_filter_button_clicked)

# Display all widgets
display(site_dropdown, min_price_input, max_price_input, filter_button, output)
```
